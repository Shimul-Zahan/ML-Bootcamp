```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn import datasets
%matplotlib inline
```

```python
dataset = pd.read_csv('../data/healthcare.csv')
```

```python
plt.scatter(dataset.iloc[:, 0], dataset.iloc[:, 1])
```

```python
from sklearn.ensemble import IsolationForest
isolation_forest = IsolationForest(
    contamination=0.2  # threshold for this annomaly score
)

isolation_forest.fit(dataset)
prediction = isolation_forest.predict(dataset)

''' 
class IsolationForest(
    *,
    n_estimators: Int = 100,
    max_samples: float | int | Literal['auto'] = "auto",
    contamination: float | str = "auto",
    max_features: float | int = 1,
    bootstrap: bool = False,
    n_jobs: Int | None = None,
    random_state: Int | RandomS
    verbose: Int = 0,
    warm_start: bool = False
)
'''
```

```python
x = dataset.values
annomali_index = np.where(prediction < 0)
plt.scatter(dataset.iloc[:, 0], dataset.iloc[:, 1])
plt.scatter(x[annomali_index, 0], x[annomali_index, 1], edgecolors='g')
```
