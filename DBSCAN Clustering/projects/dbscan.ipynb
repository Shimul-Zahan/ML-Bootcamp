```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons,make_circles, make_swiss_roll, make_blobs
from sklearn.cluster import DBSCAN
%matplotlib inline
```

```python
X,y=make_moons(n_samples=500, noise=0.05)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X)
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
dbscan = DBSCAN(eps=0.3).fit(X_train)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
```

```python
# Generate complex shape data
X1, y1 = make_blobs(n_samples=300, centers=4, cluster_std=0.5, random_state=42)
X2, y2 = make_circles(n_samples=300, factor=0.3, noise=0.05)
X3, _ = make_swiss_roll(n_samples=500, noise=0.1)

# Plot the datasets
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].scatter(X1[:, 0], X1[:, 1], c=y1, cmap='viridis')
axs[0].set_title("Blobs with 4 clusters")

axs[1].scatter(X2[:, 0], X2[:, 1], c=y2, cmap='coolwarm')
axs[1].set_title("Nested Circles")

axs[2].scatter(X3[:, 0], X3[:, 2], c=X3[:, 1], cmap='plasma')
axs[2].set_title("Swiss Roll")

plt.show()
```
