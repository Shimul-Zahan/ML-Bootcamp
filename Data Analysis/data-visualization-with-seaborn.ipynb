```python
import seaborn as sns
```

```python
# Basic plotting with searborn
# seaborn default dataset
tips = sns.load_dataset('tips')
tips.head(10)
```

```python
# create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("scatterplot using seaborn")
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
# line plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("line plot using seaborn")
plt.xlabel('size')
plt.ylabel('total_bill')
plt.show()
```

```python
# plots categorical variable
# bar plots
sns.barplot(x='day',y='total_bill', data=tips)
plt.title("bar plot using categorical variable")
plt.xlabel('day')
plt.ylabel('total_bill')
plt.show()
```

```python
# Box plot
sns.boxplot(x='day',y='total_bill', data=tips)
plt.title("box plot using categorical variable")
plt.xlabel('day')
plt.ylabel('total_bill')
plt.show()
```

```python
# violin plot
sns.violinplot(x='day',y='total_bill', data=tips)
plt.title("violin plot using categorical variable")
plt.xlabel('day')
plt.ylabel('total_bill')
plt.show()
```

```python
# histogram
sns.histplot(tips['total_bill'], bins=10, kde=True)
```

```python
# kde plot
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
# pair plots
sns.pairplot(tips)
```

```python
# heatmap
corr=tips[['total_bill', 'tip', 'size']].corr()
corr
# tips.head()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
# customized the plots
import pandas as pd
data = pd.read_csv('sales_data.csv')
data.head()
```

```python
# plots total sales by product
plt.figure(figsize=(10, 8))
sns.barplot(x='Product Category', y='Total Revenue', data=data, estimator=sum) #estimator for aggrigations
plt.title("Total Sales by Product")
plt.xlabel('Product')
plt.ylabel('Total revenue')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='Region', y='Total Revenue', data=data, estimator=sum) #estimator for aggrigations
plt.title("Total Sales by Region")
plt.xlabel('Product')
plt.ylabel('Total revenue')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='Payment Method', y='Total Revenue', data=data, estimator=sum) #estimator for aggrigations
plt.title("Total Sales by Payment Method")
plt.xlabel('Product')
plt.ylabel('Total revenue')
plt.show()
```
