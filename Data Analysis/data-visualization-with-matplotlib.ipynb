```python
# Here is the heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df = pd.DataFrame(data)

# Create a heatmap
plt.figure(figsize=(8, 6))  # Optional: set the size of the plot
sns.heatmap(df, annot=True, cmap='YlGnBu', linewidths=0.5)

# Show the plot
plt.title('Sample Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for heatmap (sales over days and weeks)
data = np.array([[10, 20, 30],  # Week 1 sales for 3 days
                 [40, 120, 60],  # Week 2 sales
                 [70, 80, 111]]) # Week 3 sales

# Display heatmap
plt.imshow(data, cmap='coolwarm')  # coolwarm is a color map
plt.colorbar()  # Show color legend
plt.title('Sales Heatmap')
plt.xlabel('Day of the Week')
plt.ylabel('Week')
plt.show()
```

```python
# No need for now

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Define the cost function
def cost_function(w1, w2):
    # A simple quadratic cost function
    return w1**2 + w2**2

# Step 2: Create a grid of w1 and w2 values
w1 = np.linspace(-10, 10, 100)  # range of values for w1
w2 = np.linspace(-10, 10, 100)  # range of values for w2
W1, W2 = np.meshgrid(w1, w2)
Z = cost_function(W1, W2)  # Cost for each combination of w1 and w2

# Step 3: Plot 3D surface and 2D heatmap (contour)
fig = plt.figure(figsize=(14, 8))

# 3D Surface plot
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax1.plot_surface(W1, W2, Z, cmap='viridis', alpha=0.8)
ax1.set_xlabel('w1')
ax1.set_ylabel('w2')
ax1.set_zlabel('Cost')
ax1.set_title('3D Cost Function Surface')
fig.colorbar(surf, ax=ax1)

# 2D Contour plot (heatmap view)
ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contourf(W1, W2, Z, cmap='coolwarm')
plt.colorbar(contour, ax=ax2)
ax2.set_xlabel('w1')
ax2.set_ylabel('w2')
ax2.set_title('Gradient Descent Heatmap (Cost)')

# Simulating gradient descent steps
# Starting point for gradient descent
w1_start, w2_start = 8, 8
alpha = 0.1  # Learning rate
points_w1, points_w2 = [w1_start], [w2_start]  # To store path
for _ in range(10):  # 10 steps
    grad_w1 = 2 * points_w1[-1]
    grad_w2 = 2 * points_w2[-1]
    next_w1 = points_w1[-1] - alpha * grad_w1
    next_w2 = points_w2[-1] - alpha * grad_w2
    points_w1.append(next_w1)
    points_w2.append(next_w2)

# Plotting gradient descent steps on contour plot
ax2.plot(points_w1, points_w2, 'ro-', label='Gradient Descent Steps')
ax2.legend()

plt.tight_layout()
plt.show()
```
