```python
### handling missing values
df.isnull().any()
```

```python
# get the sum of missing value
df.isnull().sum()
```

```python
df.filled = df.fillna(0)
```

```python
# Filling missing value with the mean value of this col
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
df['Value'] = df['Value'].fillna(df['Value'].mean())
```

```python
df
```

```python
df.isnull().sum()
```

```python
# How we calculate the mean value
df['Sales'].sum() / df['Sales'].count()
```

```python
# renaming column
# df = df.rename(columns={'column name': 'new name'})

df = df.rename(columns={'Date': 'Sales Date'})
df.head()
```

```python
df['Value'] = df['Value'].apply(lambda x: x + (x / 100))
```

```python
# data aggreagation and grouping
grouped_meand = df.groupby(['Region','Category', 'Sales Date'])['Value'].max()
grouped_meand
```
