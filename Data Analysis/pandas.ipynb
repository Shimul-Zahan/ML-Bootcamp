```python
# Heres is the introduction of pandas
import pandas as pd
import numpy as np
```

```python
## series
## pd series is a 1D array like a one single collum
data = np.arange(1, 6)
series = pd.Series(data)
series
```

```python
# create a series fro dictionary
dict = {
    "a": 1, "b": 2, "c": 3
}

series_dict = pd.Series(dict)
series_dict
```

```python
### dataframe
# df or dataframe is 2D array or table
# Create a dataframe from dictionary of list
data = {
    "name": ["Shimul", "Eleush", "Zahan"],
    "age":[20, 30, 35],
    "cities": ["Rajshahi", "Dhaka", "Khulna"],
    "Jobs": ["Engineer", "Doctor", "Student"],
}
df = pd.DataFrame(data)
df
```

```python
# create this df to numpy array
np.array(df)
```
