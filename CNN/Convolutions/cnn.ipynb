```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = np.array(
    [
        [0, 0 , 0, 255, 255, 255],
        [0, 0 , 0, 255, 255, 255],
        [0, 0 , 0, 255, 255, 255],
        [0, 0 , 0, 255, 255, 255],
        [0, 0 , 0, 255, 255, 255],
        [0, 0 , 0, 255, 255, 255],
    ]
)

# here we normalize the image
image = image / 255
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(image, cmap='gray')
plt.title("Gray Scale image representation")
plt.axis("off")
plt.show()
```

```python
# define a kernel and convolve it
vertical_filter = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1],
])
```

```python
# Defines all the sizes
image_height, image_width = image.shape
kernel_size = vertical_filter.shape[0]
output_image_height = image_height - kernel_size + 1
output_image_width = image_width - kernel_size + 1

output_image = np.zeros((output_image_height, output_image_width))

# output_image

for i in range(output_image_height):
    for j in range(output_image_width):
        region = image[i:i+kernel_size, j:j+kernel_size]
        print(region)
        output_image[i, j] = np.sum(region * vertical_filter)
        
# Display the original and convolved images
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
# plt.axis("off")

# Convolved image
plt.subplot(1, 2, 2)
plt.imshow(output_image, cmap="gray")
plt.title("Convolved Image (Manual)")
# plt.axis("off")

plt.show()
```
