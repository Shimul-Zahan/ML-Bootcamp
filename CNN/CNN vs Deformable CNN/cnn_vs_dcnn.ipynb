```python
# ছবি লোড করা
image = cv2.imread('./resources/9919.png')

# চেক করা ছবি লোড হয়েছে কিনা
if image is None:
    print("Error: Image not found!")
else:
    # BGR থেকে RGB এবং Grayscale কনভার্সন
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Matplotlib দিয়ে দেখানো
    plt.figure(figsize=(10,5))

    # RGB ইমেজ দেখানো
    plt.subplot(1,2,1)
    plt.imshow(image_rgb)
    plt.title("RGB Image")
    plt.axis('off')

    # Grayscale ইমেজ দেখানো
    plt.subplot(1,2,2)
    plt.imshow(image_gray, cmap='gray')
    plt.title("Grayscale Image")
    plt.axis('off')

    # Show both images
    plt.show()
```
