```python
import torch
import torch
import torch.nn as nn
import torchvision.ops as ops

# Input Image (Batch Size=1, Channels=1, Height=5, Width=5)
image = torch.randn(1, 1, 5, 5)
print(image)

# Define a Deformable Convolution Layer
deform_conv = ops.DeformConv2d(in_channels=1, out_channels=1, kernel_size=3, padding=1, bias=False)

print(deform_conv)

# Generate Random Offsets (Batch Size=1, 2*3*3=18, Height=5, Width=5)
offsets = torch.randn(1, 18, 5, 5)

print(offsets)

# Apply Deformable Convolution
output = deform_conv(image, offsets)

print("Output Shape:", output.shape)
```
