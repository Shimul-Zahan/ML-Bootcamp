```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
test = pd.read_csv('../Data/titanic/test.csv')
train = pd.read_csv('../Data/titanic/train.csv')
```

```python
train.head()
test.shape
train.shape
train.info()
test.info()
```

```python
# train.isnull().sum()
# test.isnull().sum()
train.dropna(inplace=True)
test.dropna(inplace=True)
```

```python
# cat_cols

train.head(2)

test_id = test['PassengerId']

train.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
test.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# train_copy = train.copy()
# test_copy = test.copy()
```

```python
train.head()
cat_cols = train.select_dtypes(include='object').columns
# cat_cols
# test_id

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for col in cat_cols:
    train[col] = encoder.fit_transform(train[col])
    test[col] = encoder.transform(test[col])
    
    
# train[cat_cols] = train[cat_cols].apply(lambda col: encoder.fit_transform(col))
```

```python
# train.describe()

numeric_cols = train.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(train[col], kde=True)
    plt.title(col)
    plt.show()
```
